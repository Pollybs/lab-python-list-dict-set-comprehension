````python

products = ["t-shirt", "mug", "hat", "book", "keychain"]
customer_orders = []
number_orders_list = []


def initialize_inventory(products):
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory


def get_customer_orders():
    num_orders = 0

    while True:
        enter_num_orders = input("Enter the number of customer orders: ")
        if enter_num_orders.isdigit():
            num_orders = int(enter_num_orders)
            break
        else:
            print("Invalid input. Please enter a valid digit.")

    customer_orders = [input(f"Enter the name of product {i + 1}: ") for i in range(num_orders)]
    print(f"Number of orders: {num_orders}")
    print("Customer Orders:")
    print('\n'.join(customer_orders))
    return customer_orders


def calculate_total_price(customer_orders):
    prices = sum([float(input(f"Enter the price of product {product}: ")) for product in customer_orders])
    print("Total Price of customer order:", prices)
    return prices


inventory = initialize_inventory(products)

print("Current inventory")
print('\n'.join([f"{product}: {quantity}" for product, quantity in inventory.items()]))

customer_orders = get_customer_orders()
prices = calculate_total_price(customer_orders)

inventory_updated = {product: quantity - 1 if product in customer_orders and quantity > 0 else quantity for product, quantity in inventory.items()}
inventory_updated = {product: quantity for product, quantity in inventory_updated.items() if quantity > 0}

print("\nInventory updated:")
print('\n'.join([f"{product}: {quantity}" for product, quantity in inventory_updated.items()]))
````
